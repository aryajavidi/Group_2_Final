```python
all_amenities=la_data['amenities'].str.replace('{', '').str.replace('}', '').str.replace('"', '').str.cat(sep=',')
amenities_list=all_amenities.split(',')
amenities_dict={}
for each_amenity in amenities_list: 
    if each_amenity in amenities_dict: 
        amenities_dict[each_amenity]+=1
    else: 
        amenities_dict[each_amenity]=1
amenities_dict
```

```python
def sort_dict_by_value(d, reverse = False):
    return dict(sorted(d.items(), key = lambda x: x[1], reverse = reverse))

sort_dict_by_value(amenities_dict, reverse=True)
```

```python
amenities_type=list(amenities_dict.keys())
amenities_type_price={}
for each_amenity in amenities_type: 
    # each_amenity=amenities_type[0]
    amenities_type_price[each_amenity]=la_data[la_data['amenities'].str.contains(each_amenity)]['log_price'].mean()
    
amenities_type_price
```

```python
sorted_amen_dict = sort_dict_by_value(amenities_type_price, reverse=True)
sorted_amen_dict
```
