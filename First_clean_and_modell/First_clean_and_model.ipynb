```python
ny_data = airbnb_df[airbnb_df["city"]=="NYC"]
```

```python
ny_data.head()
```

```python
all_amenities=ny_data['amenities'].str.replace('{', '').str.replace('}', '').str.replace('"', '').str.cat(sep=',')
amenities_list=all_amenities.split(',')
amenities_dict={}
for each_amenity in amenities_list: 
    if each_amenity in amenities_dict: 
        amenities_dict[each_amenity]+=1
    else: 
        amenities_dict[each_amenity]=1
amenities_dict
```

```python
def sort_dict_by_value(d, reverse = False):
    return dict(sorted(d.items(), key = lambda x: x[1], reverse = reverse))

sort_dict_by_value(amenities_dict, reverse=True)
```

```python
amenities_type=list(amenities_dict.keys())
amenities_type_price={}
for each_amenity in amenities_type: 
    # each_amenity=amenities_type[0]
    amenities_type_price[each_amenity]=ny_data[ny_data['amenities'].str.contains(each_amenity)]['log_price'].mean()
    
amenities_type_price
```

```python
sort_dict_by_value(amenities_type_price, reverse=True)
```

```python
def has_amenity(amenity_list, amenity): 
    if amenity in amenity_list: 
        return True
    else: 
        return False
```

```python
ny_data['has_pool']=ny_data['amenities'].str.lower().apply(lambda x: has_amenity(x, 'pool'))
```

```python
ny_data['has_fireplace']=ny_data['amenities'].str.lower().apply(lambda x: has_amenity(x, 'indoor fireplace'))
```

```python
ny_data['has_wifi']=ny_data['amenities'].str.lower().apply(lambda x: has_amenity(x, 'wireless_internet'))
```

```python
ny_data['has_kitchen']=ny_data['amenities'].str.lower().apply(lambda x: has_amenity(x, 'kitchen'))
```

```python
ny_data['has_dishwasher']=ny_data['amenities'].str.lower().apply(lambda x: has_amenity(x, 'dishwasher'))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
# reg = LinearRegression().fit(X_train, y_train)
reg = RandomForestRegressor().fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```
