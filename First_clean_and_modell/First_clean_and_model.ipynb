```python
ny_data = airbnb_df[airbnb_df["city"]=="NYC"]
```

```python
ny_data_cleaned = ny_data[["log_price", "room_type","accommodates","neighbourhood","zipcode","bedrooms","bathrooms"]]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df2 = ny_data_cleaned.copy()
df2['room_type'] = le.fit_transform(df2['room_type']) 
df2.head()
```

```python
df3 = df2.copy()
df3['neighbourhood'] = le.fit_transform(df2['neighbourhood']) 
df3.head()
```

```python
df4 = df3.copy()
df4['zipcode'] = le.fit_transform(df2['zipcode']) 
df4.head()
```

```python
df4.dtypes
```

```python
df4.isnull().sum()
```

```python
df4.dropna(inplace=True)
```

```python
df4.isnull().sum()
```

```python
df4.head()
```

df4.isnull()

```python
X = df4.drop("log_price", axis=1)
y= df4["log_price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```

```python
print(mse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

print("R-squared value:", r2)
```
